importing necessary libraries

```python
import pandas as pd
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

reading the table 

```python
df=pd.read_html(url)
```

table to dataframe

```python
df1=df[0]
```

```python
df1.head(20)
```

Ignoring cells with a borough that is Not assigned.

```python
df1['Borough'].replace('Not assigned', np.nan, inplace=True)
df1.dropna(subset=['Borough'], inplace=True)
```

```python
df1.head()
```

Aggregating

```python
df1 = df1.groupby(['Postal Code', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
df1.head(50)
```

```python
df1.shape
```

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')
df_geo.head()
```

Adding latitude and longitude

```python
df2 = pd.merge(df1, df_geo, on=['Postal Code'], how='inner')
df2.head()
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
from geopy.geocoders import Nominatim
import folium
```

Exploring neighborhoods in Toronto

```python
address = 'Toronto, Canada'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

Visualising Toronto neighborhoods

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for borough, neighborhood, lat, long in zip( df2['Borough'], df2['Neighborhood'],df2['Latitude'], df2['Longitude']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, long],
        radius=3,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
